```python
import FK_stats_module

numbers = []

while True:
    print("\nMenu:")
    print("1. Enter numbers")
    print("2. Mean")
    print("3. Variance")
    print("4. Standard Deviation")
    print("5. Standard Error")
    print("6. Z-score")
    print("7. Show Summary")
    print("8. End Session")

    choice = input("Pick from the menu ")

    if choice == '1':
        num = input("Input a number and type 'done' when finished: ")
        while num.lower() != 'done':
            try:
                numbers.append(float(num))
            except Error:
                print("ERROR.")
            num = input("Input a number and type 'done' when finished:")
  
    elif choice == '2':
        if numbers:
            mean = FK_stats_module.calculate_mean(numbers)
            print(f"Mean is {mean}")
        else:
            print("Please enter data in option 1!")

    elif choice == '3':
        if len(numbers) >= 2:
            variance = FK_stats_module.calculate_variance(numbers)
            print(f"Variance is {variance}")
        else:
            print("Please enter at least two numbers")

    elif choice == '4':
        if numbers:
            std_deviation = FK_stats_module.calculate_standard_deviation(numbers)
            print(f"Standard Deviation is {std_deviation}")
        else:
            print("Please enter data in option 1!")

    elif choice == '5':
        if numbers:
            std_error = FK_stats_module.calculate_standard_error(numbers)
            print(f"Standard Error is {std_error}")
        else:
            print("Please enter data in option 1!")

    elif choice == '6':
        if numbers:
            newNumber = float(input("Enter a new number: "))
            z_score = FK_stats_module.calculate_z_score(newNumber, numbers)
            print(f"Z-score is {z_score}")
        else:
            print("Please enter data in option 1!")

    elif choice == '7':
        if numbers:
            mean = FK_stats_module.calculate_mean(numbers)
            variance = FK_stats_module.calculate_variance(numbers)
            std_deviation = FK_stats_module.calculate_standard_deviation(numbers)
            std_error = FK_stats_module.calculate_standard_error(numbers)
            print(f"Summary:")
            print(f"Mean is {mean}")
            print(f"Variance: {variance}")
            print(f"Standard Deviation is {std_deviation}")
            print(f"Standard Error is {std_error}")
        else:
            print("Please enter data in option 1!")

#not sure if this is the right thing...I tried my best to stop it from repeating
    elif choice == '8':
        break
    
    else:
        print("Please select an option from the menu")
print("Thank you and have a wonderful day! :)")
```
